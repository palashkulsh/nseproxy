```python
import pandas as pd

def get_ltp(symbol="NIFTY-I"):
    hist_data = td_obj.get_historic_data(symbol, duration='3 d',bar_size="2 min")
    ltp = hist_data[-1]['c']
    return ltp

get_ltp()

def get_hist_data(symbol="NIFTY-I", duration="3 d", bar_size="2 min"):
    hist_data = td_obj.get_historic_data(symbol, duration=duration,bar_size=bar_size)
    return pd.DataFrame(hist_data)
```

### get current expiry

```python
from datetime import datetime , timedelta
from dateutil.relativedelta import relativedelta, TH

def getCurrentExpiry():
    end_of_month = datetime.today() + relativedelta(day=31)
    last_thursday = end_of_month + relativedelta(weekday=TH(-1))
    expiry_date=last_thursday.strftime("%Y%m%d")
    return expiry_date

getCurrentExpiry()
```

### get option chain and strikes

```python
import requests

def getOptionChain(symbol="NIFTY",expiry=getCurrentExpiry()):
    option_chain=pd.read_csv('https://api.truedata.in/getOptionChain?user=FYERS872&password=fjt3KN3s&symbol='+symbol+'&expiry=20210624&csv=true',names=["id","symbol","type","empty","exchange","something","strike","expiry","symbol2","symbol3"])
    return option_chain

#getOptionChain()

def getStrikes(symbol, op_symbol, expiry=getCurrentExpiry(), percent_strikes=5):
    opc_data= getOptionChain(symbol=op_symbol, expiry=expiry)
    ltp= get_ltp(symbol=symbol)
    print(ltp, percent_strikes)
    #opc_data_sorted = opc_data.sort_values(by=["strike"])
    filtered_strikes=opc_data.query('strike>(1-(@percent_strikes/100))*@ltp and strike<(1+ (@percent_strikes/100))*@ltp')
    return filtered_strikes

getStrikes(symbol="NIFTY-I", op_symbol="NIFTY",percent_strikes=3, expiry=getCurrentExpiry())
```

### get historical Data for strikes

```python
import time

def getData(symbol="NIFTY-I", op_symbol="NIFTY",percent_strikes=3, duration="7 D", bar_size="5 min"):
    opc_data = getStrikes(symbol="NIFTY-I", op_symbol="NIFTY",percent_strikes=3,expiry=getCurrentExpiry())
    put_strikes = opc_data.query('type=="PE"').symbol
    call_strikes= opc_data.query('type=="CE"').symbol
    all_dfs=[]
    strikes= pd.concat([put_strikes,call_strikes,pd.Series(['NIFTY-I'])]).values
    for strike in strikes:
        print(strike)
        all_dfs.append(get_hist_data(symbol=strike, duration=duration, bar_size=bar_size)[["time","c","v","oi"]])
        time.sleep(0.5)
        
    # merging all strike data
    merged_data = pd.concat([d.set_index("time") for d in all_dfs], keys=strikes, join="outer",axis=1)
    return merged_data
    
merged_data = getData(duration="17 D",percent_strikes=3)
```
